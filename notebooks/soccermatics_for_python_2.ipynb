```python
#Function to draw the pitch
import matplotlib.pyplot as plt
import numpy as np
import sys
%matplotlib inline
sys.path.append("/mnt/d/xFootball/SoccermaticsForPython/")#os.getcwd() + '/modules')
from config import STATSBOMB_DATA
```

```python
#Size of the pitch in yards (!!!)
pitchLengthX=120
pitchWidthY=80
```

```python
#ID for England vs Sweden Womens World Cup
match_id_required = '69301'
home_team_required ="England Women's"
away_team_required ="Sweden Women's"
file_name= f"{match_id_required}.json"
```

```python
#Load in all match events 
import json
with open(f'{STATSBOMB_DATA}/events/{match_id_required}.json') as data_file:
    #print (mypath+'events/'+file)
    data = json.load(data_file)
```

```python
#get the nested structure into a dataframe 
#store the dataframe in a dictionary with the match id as key (remove '.json' from string)
from pandas.io.json import json_normalize
df = json_normalize(data, sep = "_").assign(match_id = file_name[:-5])
df.head()
```

```python
#A dataframe of shots
shots = df.loc[df['type_name'] == 'Shot'].set_index('id')
shots.head()
```

```python
#Draw the pitch
from FCPython import createPitch
(fig,ax) = createPitch(pitchLengthX,pitchWidthY,'yards','gray')

#Plot the shots
for i,shot in shots.iterrows():
    x=shot['location'][0]
    y=shot['location'][1]
    
    goal=shot['shot_outcome_name']=='Goal'
    team_name=shot['team_name']
    
    circleSize=2
    #circleSize=np.sqrt(shot['shot_statsbomb_xg'])*12

    if (team_name==home_team_required):
        if goal:
            shotCircle=plt.Circle((x,pitchWidthY-y),circleSize,color="red")
            plt.text((x+1),pitchWidthY-y+1,shot['player_name']) 
        else:
            shotCircle=plt.Circle((x,pitchWidthY-y),circleSize,color="red")     
            shotCircle.set_alpha(.2)
    elif (team_name==away_team_required):
        if goal:
            shotCircle=plt.Circle((pitchLengthX-x,y),circleSize,color="blue") 
            plt.text((pitchLengthX-x+1),y+1,shot['player_name']) 
        else:
            shotCircle=plt.Circle((pitchLengthX-x,y),circleSize,color="blue")      
            shotCircle.set_alpha(.2)
    ax.add_patch(shotCircle)
    
    
plt.text(5,75,away_team_required + ' shots') 
plt.text(80,75,home_team_required + ' shots') 
     
fig.set_size_inches(10, 7)
plt.show()
```

```python
from mplsoccer.pitch import Pitch
pitch = Pitch(pitch_type='statsbomb', pitch_color='#374C32', line_color='white',stripe=False,figsize=(10, 7))
fig, ax = pitch.draw()
shot_alpha = lambda x: 1 if x=='Goal' else 0.2
shot_color = lambda x: 'red' if x==home_team_required else 'orange'
flip_x_if_away = lambda x: x if shot.team_name==home_team_required else pitchLengthX-x
flip_y_if_away = lambda y: y if shot.team_name==home_team_required else pitchWidthY-y

for shot in shots.itertuples():
    shot_plot = pitch.scatter(flip_x_if_away(shot.location[0]),flip_y_if_away(shot.location[1]), 
                                s=100, ax=ax, edgecolor='white',zorder=2,
                                color=shot_color(shot.team_name),
                                alpha=shot_alpha(shot.shot_outcome_name))
    if shot.shot_outcome_name=='Goal':
        ax.text(flip_x_if_away(shot.location[0]+1),flip_y_if_away(shot.location[1]+1),shot.player_name, color='white') 

ax.text(5,75,away_team_required + ' shots') 
ax.text(80,75,home_team_required + ' shots')

```

```python
import pandas as pd
import glob
import warnings
warnings.filterwarnings('ignore')
pd.set_option('display.max_columns', None)
```

```python
# get all FRA matches
# get possession chains - their durations 
# all giroud successful receives in final 10s of possession chain
# chain end clf - shot/disposs/goal etc.
```

```python
competitions = pd.read_json(f"{STATSBOMB_DATA}/competitions.json")
wc18 = str(competitions.query("country_name=='International' and competition_gender=='male'")['competition_id'].values[0])
# glob.glob(f'{STATSBOMB_DATA}/matches/{wc18}/*')


matches = pd.DataFrame()
for MATCH_FILE in glob.glob(f'{STATSBOMB_DATA}/matches/{wc18}/*'):
    # matches = pd.concat([matches, pd.read_json(MATCH_FILE)])
    matches = pd.concat([matches, pd.json_normalize(json.load(open(MATCH_FILE)), sep='_')])
matches = matches[(matches['home_team_country_name'] == 'France') | (matches['away_team_country_name'] == 'France')].sort_values('match_date')

events = pd.DataFrame()
for matchID in matches.match_id:
    match = pd.json_normalize(json.load(open(f"{STATSBOMB_DATA}/events/{matchID}.json")), sep='_')
    match['match_id'] = matchID
    events = pd.concat([events,match.reset_index(drop=True)]).reset_index(drop=True)
    print (events.shape)
events.reset_index(drop=True).index

```

```python
','.join(events.columns.to_list())
```

```python
events.query("possession_team_name=='France'").groupby(['match_id','possession','timestamp'])['type_name'].apply(list)
```

```python
for i,possession_df in events.query("possession_team_name=='France'").groupby(['match_id','possession']):
    if any(possession_df.type_name.str.contains('Half')): continue
    for i, row in possession_df.iterrows():
        possession_df.at[i,'chain_so_far'] =  (pd.Timedelta(row.timestamp) - pd.Timedelta(possession_df.timestamp.iloc[0])).seconds
        possession_df.at[i,'chain_left'] =  (pd.Timedelta(possession_df.timestamp.iloc[-1])-pd.Timedelta(row.timestamp)).seconds
    possession_10 = possession_df.query("chain_left<=10")    
    if any(possession_10.type_name.str.contains('Shot')) & any(possession_10.player_name.str.contains('Giroud')):
        print (possession_10[['minute','second','player_name','type_name']])
    else:
        continue
    print ('=='*5)
    # break
```
